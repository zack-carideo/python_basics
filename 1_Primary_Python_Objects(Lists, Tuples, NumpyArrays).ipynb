## Lists 
* holds data in order it was added
* mutable 
* index 
* slicing requires start and end positions, the start position is included in output, the end position is not list[start:end]

##### NOTE: when a list is copied into another list, the copied list will point to the original, so any changes made to the copied list will also be made to to original list! To avoid this, and prevent changes int he copied list to take affect on the original list , you must explictly copy the original list using the list() or [:] syntax

```python
#example list
Names = ['zack','jim','george','mike']
Ages = ['3','45','324','5']

##Change a list element 
Names[0]='Zack'

##Add element to list 
Name_Ext = Names + ['Jack','Jill']

#Delete element from list 
del(Name_Ext[len(Name_Ext)-1])
del(Ages[len(Ages)-1])

##Append to list
Name_Ext.append('zachary')

#identify index of a value 
position = Names.index('Zack')
Names[position]

#remove element from a list and save the element removed to an object 
Myname = Names.pop(position)

#sort a list 
#sorted() function returns a new list and does not affect the list you passed into the function
sortedlist = sorted(Names)

##combine lists 
#Note: Explictly coping each list into the new lsit to prevent changes in new list from perferating down to original list
NamesAndAges = list(Names) + list(Ages)

#####
##LIST OF LISTS 
#####
#Create list of list from 2 lists
People = [Names,Ages]

#subset list of lists 
People[1] #position of list you want to work with
People[0][:2] #position of inner elmement of inner list you want to return 
```

## NUMPY
* Enables you to process entire columns with simple syntax, an alternative to python list 
* Can perform calculations across entire arrays in elementwise fashion 
* Applicable on numeric vectors
* Can only contain one type, if multiple types of data are included in a numpy array, all inputs will be converted to strings
* When T/F is part of a array, True == 1 and False == 0.

##### Note: The typical arithmetic operators, such as +, -, * and / have a different meaning for regular Python lists and numpy arrays

```python
import numpy as np

#NUMERIC VECTORS TO OPERATE WITH 
TransAmount = np.array(df['TRANSACTION_AMOUNT'])
LedgerBal = np.array(df['BALANCE_LEDGER'])

#output type of object iterating over 
print(type(TransAmount))

#subset numpy array 
print(TransAmount[1])

#subset numpy array with logical conditions 
print(TransAmount> 20) #output logical array stating weather or not condition is true 
print(TransAmount[TransAmount>20])  #subset original array based on logical condition 

#element wise sum of numpy arrays 
#Note: using + with Traditional lists will append elements from the first list to the second
#, with numpy arrays, the values in one array are added to the other array  
NewBal = LedgerBal - TransAmount

##Boolean Operators with Numpy 
print(np.logical_or(LedgerBal > 10, LedgerBal>1000))
print(np.logical_and(LedgerBal > 37000000, TransAmount<1000))
```

## 2D Numpy Arrays(Basic Navigation)
* you can create a 2d numpy array from a list of lists 

```python
#2 1D arrays 
array1 = np.array(list(LedgerBal))
array2 = np.array(list(TransAmount))

#2D Array(Option 1)
np_2d = np.array([array1,array2])
np_2D = np.transpose(np_2d)

#2D Array(Option 2)
np_2d = np.column_stack((array1,array2))

#look at characteristics of array
print(array1.shape,array2.shape)
print(type(np_2d))
print(np_2d.shape)
print(np_2D.shape)

#selecting specific information from the arrays 
np_2d[3][0] #select the fourth element from the first row 
np_2d[3,0]

#select both rows and the first 3 columns 
np_2d[:][1:3]

#select entire second row and first 3 columns 
np_2d[1,1:3]
```

### NUMPY ARRAY STATS 

```python
##BASIC STATS 
print(np.mean(np_2D[:,1]))   #mean of all rows in first column (LedgarBalance)
print(np.median(np_2D[:,1])) #median of all rows in first column (LedgarBalance)
print(np.std(np_2D[:,1])) #std of all rows in first column (LedgarBalance)
corr = np.corrcoef(np_2D[:,0],np_2D[:,1]) #Corrolation between two arrays in 2d array
print("Correlation: " + str(corr))
```
