```python
#Import Data 
#Option 2: Load data into pandas df, which can then be converted into any other data type 
import pandas as pd
df_pop = pd.read_csv('/home/docadmin/ZackC/Python Misc/Data/PpopulationData.csv', delimiter=',')
df_stats = pd.read_csv('/home/docadmin/ZackC/Python Misc/Data/WorldDebtData.csv', delimiter=',')

#column headers and dataset size
print(list(df_stats),df_stats.size)
print(list(df_pop))
```

### Subsetting Panda's Dataframes 

```python
#subset pandas to only include non empty info 
df_stats_subset = df_stats[df_stats['2017Q2'].notnull() & (df_stats['2005Q4'].notnull())]

#rename column from population dataset you want to joing on 
df_pop=df_pop.rename(columns = {'Country ':'Country Code'})

#select the first 4 columns of the population dataset 
df_pop_subset = df_pop.iloc[:,[0,1,3]]
print(list(df_pop_subset),list(df_stats_subset))
```

### Joinging / Merginging Pandas Dataframes 

* **Syntax:** pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,left_index=False, right_index=False, sort=True,suffixes=('_x', '_y'), copy=True, indicator=False,validate=None)



```python
#Merge population and population debt statistics into a single dataframe based on country code 
PopAndStats = pd.merge(df_stats_subset, df_pop_subset, how='left', on='Country Code', sort=True)


#set row index's for each element in list 
PopAndStats.index = [PopAndStats['Country Code']]
```
